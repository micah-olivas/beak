## Test 2: Interactive Remote Session with nest()

The interactive approach provides more control and progress tracking during the nest operation.

```python
# Test interactive remote session with nest functionality
try:
    print("Creating interactive remote session...")
    
    # Connect to remote server (will prompt for credentials if needed)
    # session = connect_remote(user='your_username', host='shr-zion.stanford.edu')
    session = connect_remote()  # Uses default host
    
    print("✅ Connected to remote server!")
    
    # Run interactive nest function
    print("\nRunning interactive nest...")
    session.nest_interactive()
    
    print("✅ Interactive nest completed!")
    
    # You can now use the session for other operations
    print(f"Session info: Connected to {session.host}")
    
except Exception as e:
    print(f"❌ Error with interactive session: {e}")
    print("This is expected if you don't have remote access configured")
```
